```python
# function to add value labels
def addlabels(x,y,offset):
    for i in range(len(x)):
        plt.text(x[i], y[i]+offset, f"{y[i]:.2f}", ha = 'center',
                 bbox = dict(facecolor = 'white', alpha =.8))
```

```python
# snr-rmse improvement
snr_list = [-20,-15,-10,-5,0,5,10,15,20,25,30]

transformer_rmse = np.load('../../../results\experiment_1/scenario_2/transformer_rmse1.npy')
music_rmse = np.load('../../../results\experiment_1/scenario_2/music_rmse1.npy')
capon_rmse = np.load('../../../results\experiment_1/scenario_2/capon_rmse1.npy')
l1_svd_rmse = np.load('../../../results\experiment_1/scenario_2/l1-svd_rmse1.npy')
cnn_papageorgiou2021_rmse = np.load('../../../results\experiment_1/scenario_2/cnn_papageorgiou2021_rmse1.npy')
dnn_pavel2021_rmse = np.load('../../../results\experiment_1/scenario_2/dnn_pavel2021_rmse1.npy')
dae_papageorgiou2020_rmse = np.load('../../../results\experiment_1/scenario_2/dae_papageorgiou2020_rmse1.npy')
cnn_wu2022_rmse = np.load('../../../results\experiment_1/scenario_2/cnn_wu2022_rmse1.npy')

lowest_errors = []
for i in range(len(snr_list)):
    error = np.min([music_rmse[i], capon_rmse[i], l1_svd_rmse[i], cnn_papageorgiou2021_rmse[i], dnn_pavel2021_rmse[i], dae_papageorgiou2020_rmse[i]])
    lowest_errors.append(error)

error_diff = np.array(transformer_rmse-lowest_errors)
snr_list = np.array(snr_list)
    
fig = plt.figure(figsize=(10,5))
plt.bar(snr_list[error_diff<0], error_diff[error_diff<0], width=3, color="green", edgecolor="black", linewidth=2.7)
plt.bar(snr_list[error_diff>0], error_diff[error_diff>0], width=3, color="red", edgecolor="black", linewidth=2.7)
addlabels(snr_list[error_diff<0], error_diff[error_diff<0], -0.05)
addlabels(snr_list[error_diff>0], error_diff[error_diff>0], 0.03)
plt.title("RMSE Improvement of the Proposed Method Compared to the Lowest RMSE Among Others")
plt.xlabel("SNR (dB)")
plt.ylabel("$\Delta$RMSE ($^\circ$)")
plt.ylim([-0.8,0.1])
plt.grid()
```
