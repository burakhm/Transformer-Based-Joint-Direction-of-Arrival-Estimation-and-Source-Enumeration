```python
# function to add value labels
def addlabels(x,y,offset):
    for i in range(len(x)):
        plt.text(x[i], y[i]+offset, f"{y[i]:.2f}", ha = 'center',
                 bbox = dict(facecolor = 'white', alpha =.8))
```

```python
# snr-hausdorff improvement
snr_list = [-20,-15,-10,-5,0,5,10,15,20,25,30]
# snr_list = [-10,-5,0,5,10,15,20,25,30]

transformer_dnn_hausdorff = np.load('../../../results\experiment_3/scenario_1/transformer_hausdorff1.npy')
#transformer_3sigma_snacc = np.load('../../../results\experiment_3/scenario_1/transformer_3sigma_snacc1.npy')
music_aic_hausdorff = np.load('../../../results\experiment_3/scenario_1/music_aic_hausdorff1.npy')
music_mdl_hausdorff = np.load('../../../results\experiment_3/scenario_1/music_mdl_hausdorff1.npy')
music_sorte_hausdorff = np.load('../../../results\experiment_3/scenario_1/music_sorte_hausdorff1.npy')
music_eigen_threshold_hausdorff = np.load('../../../results\experiment_3/scenario_1/music_eigen-threshold_hausdorff1.npy')
music_eigen_increment_hausdorff = np.load('../../../results\experiment_3/scenario_1/music_eigen-increment_hausdorff1.npy')
music_areg_hausdorff = np.load('../../../results\experiment_3/scenario_1/music_areg_hausdorff1.npy')
music_t_gane_hausdorff = np.load('../../../results\experiment_3/scenario_1/music_t-gane_hausdorff1.npy')

lowest_hausdorff = []
for i in range(len(snr_list)):
    hausdorff = np.min([music_aic_hausdorff[i], music_mdl_hausdorff[i], music_sorte_hausdorff[i], music_eigen_threshold_hausdorff[i], music_eigen_increment_hausdorff[i], music_areg_hausdorff[i], music_t_gane_hausdorff[i]])
    lowest_hausdorff.append(hausdorff)

hausdorff_diff = np.array(transformer_dnn_hausdorff-lowest_hausdorff)
snr_list = np.array(snr_list)
    
fig = plt.figure(figsize=(10,5))
plt.bar(snr_list[hausdorff_diff<0], hausdorff_diff[hausdorff_diff<0], width=3, color="green", edgecolor="black", linewidth=2.7)
plt.bar(snr_list[hausdorff_diff>0], hausdorff_diff[hausdorff_diff>0], width=3, color="red", edgecolor="black", linewidth=2.7)
addlabels(snr_list[hausdorff_diff<0], hausdorff_diff[hausdorff_diff<0], -1)
addlabels(snr_list[hausdorff_diff>0], hausdorff_diff[hausdorff_diff>0], 0.6)
plt.title("Hausdorff Distance Improvement of the Proposed Method Over the Highest Accuracy Among Others")
plt.xlabel("SNR (dB)")
plt.ylabel("$\Delta$Hausdorff Distance ($^\circ$)")
plt.ylim([-15,2])
plt.grid()
```
